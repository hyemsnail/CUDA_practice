```python
%%writefile test2.cu
#include <stdio.h>

// CUDA  
__global__ void kernel() {
    printf("Hello, World from CUDA Kernel!\n");
}

int main(void) {
    kernel<<<1,1>>>();  // GPU  
    cudaDeviceSynchronize();  // GPU   
    fflush(stdout);  //   
    return 0;
}
```

```python
!nvcc -o test2 test2.cu
```

```python
!./test2
```

```python
%%writefile test3.cu
#include <stdio.h>

// CUDA  
__global__ void kernel() {
    printf("Hello, World from CUDA Kernel! Thread %d\n", threadIdx.x);
}

int main(void) {
    kernel<<<1,10>>>();  // GPU 10  
    cudaDeviceSynchronize();  // GPU   
    fflush(stdout);  //   
    printf("CUDA Kernel Execution Completed!\n");  // CPU  
    cudaDeviceReset();  // CUDA  
    return 0;
}
```

```python
!nvcc -o test3 test3.cu
```

```python
!./test3
```
